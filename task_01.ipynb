```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("titanic.csv")
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
data.head()
```

```python
Age = data["Age"]
```

```python
plt.figure( figsize=(10,5))
plt.hist( Age, bins=50)
plt.title("Titanic Age Distribution ",fontsize=20)
plt.xlabel("Age",fontsize=18)
plt.ylabel("Count",fontsize=18)
plt.grid(axis="y")
plt.show()
```

```python
Sex = data["Sex"]
```

```python
plt.figure( figsize=(10,5))
plt.hist( Sex, bins=10)
plt.title("Male And Female Survived ",fontsize=20)
plt.xlabel("Gender",fontsize=18)
plt.ylabel("Count",fontsize=18)
plt.grid(axis="y")
plt.show()
```
